```python
 #Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "../Heroes of Pymoli/resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data

count_of_players = len(purchase_data["SN"].value_counts())
total_players = pd.DataFrame({"Total Players":[count_of_players]})
total_players

   
```

```python
#define all the values and add them to a data frame 
revenue_total = (purchase_data["Price"]).sum()
average = ((purchase_data["Price"])).mean()
unique_items = len((purchase_data["Item ID"]).unique())
total_purchases = (purchase_data["Purchase ID"]).count()

#store values into a data frame
total = pd.DataFrame({"Number of Unique Items":[unique_items],
                      "Average Price":[average],
                      "Number of Purchases":[total_purchases],
                     "Total Revenue":[revenue_total]})
total
```

```python

#define all your numbers
gender = purchase_data.groupby("Gender")

count_total = gender.nunique()["SN"]

percentage = count_total/count_of_players *100
#store values into data frame
demographics = pd.DataFrame({"Total Count":count_total,
                             "Percentage of Players":percentage})
demographics_values = demographics.sort_values(["Total Count"], ascending = False)

demographics_values
```

```python
total_purchase_value = gender ["Price"].sum()

purchase_count = gender["Purchase ID"].count()

average_purchase_price = gender["Price"].mean()

average_per_person = total_purchase_value/count_total

purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Average Purchase Price":average_purchase_price,
                                 "Total Purchase Value": total_purchase_value,
                                 "Average Total Purchase Per Person":average_per_person})
purchase_analysis
```
